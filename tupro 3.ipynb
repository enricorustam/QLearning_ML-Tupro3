```python
import random
import math
import numpy as np
import pandas as pd
```

```python
f = open('dataset3.txt', 'r')
x = f.read().split('\n')
f.close()
z = []
for i in x:
    a = i.split('\t')
    z.append(a)   
r = 1
gamma = 0.8
min_step = 28 #kemungkinan minimal untuk mencapai 15x15 (500)
max_step = min_step * 20
eps = 1000 #times
```

```python
#function
def action_space(x_pos, y_pos):
    while True: #generate move dan cek keluar batas atau tidak, jika iya, maka generate move yang baru
        move = random.randint(0, 3)
        if(move == 0 and x_pos != 0):
            break
        elif(move == 1 and y_pos != 14):
            break
        elif(move == 2 and x_pos != 14):
            break
        elif(move == 3 and y_pos != 0):
            break
    return move
```

```python
def learn(t):
    return r * math.exp( -1 * (t / eps) )
```

```python
map = [] #q-table
for i in range (0, 15):
    qtab = []
    for j in range (0, 15):
        qtab2 = []
        for k in range (0, 4):
            if(i == 0 and k == 0): #jika di atas dan bergerak ke north
                pre = -9999 #menghentikan jalur keluar map dan mencegah error
            elif(i == 14 and k == 2): #jika di bawah dan bergerak ke south
                pre = -9999 
            elif(j == 0 and k == 3): #jika di kiri dan bergerak ke west
                pre = -9999 
            elif(j == 14 and k == 1):
                pre = -9999 
            else:
                pre = 0 #init zero
            qtab2.append(pre)
        qtab.append(qtab2)
    map.append(qtab)
```

```python
#Algoritma
i = 0
#waktu tempuh episode 15,15
while i < eps:
    step = random.randint(min_step,max_step) #panjang episode
    x_pos = 14 #start x
    y_pos = 0 #start y
    j = 0
    while j < step:
        randomate = random.random()
        if(randomate < learn(i)):
            move = action_space(x_pos, y_pos) #explore
        else:
            move = np.argmax(map[x_pos][y_pos]) #testing
        if(move == 0): 
            next_x = x_pos - 1 # north
            next_y = y_pos
        elif(move == 1):
            next_x = x_pos
            next_y = y_pos + 1 # east
        elif(move == 2):
            next_x = x_pos + 1 # south
            next_y = y_pos
        elif(move == 3):
            next_x = x_pos
            next_y = y_pos - 1 # west
        map[x_pos][y_pos][move] = map[x_pos][y_pos][move] + learn(i) * (int(z[next_x][next_y]) + (gamma * np.max(map[next_x][next_y])) - map[x_pos][y_pos][move])
        x_pos = next_x
        y_pos = next_y
        if(x_pos == 0 and y_pos == 14): #target tercapai, finish
            break
        j += 1
    i += 1
```

```python
x_pos = 14 #start x
y_pos = 0 #start y
score = 0
j = 0
print('map : ',map)
```

```python
while True:
    score += int(z[x_pos][y_pos])
    move = np.argmax(map[x_pos][y_pos])
    print(x_pos, y_pos) #trace bits
    if(move == 0):
        x_pos  -= 1
    elif(move == 1):
        y_pos += 1
    elif(move == 2):
        x_pos += 1
    elif(move == 3):
        y_pos -= 1
    if(x_pos == 0 and y_pos == 14): #target reached
        score += int(z[x_pos][y_pos])
        print('selesai')
        break
    j += 1
    print('gerakan',move) #trace bits
    print('score',score) #trace bits
```
